```python
import arviz as az
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
import etho_bayes as et_b
```

```python
df = pd.read_csv("resultado_bayes.csv")
```

```python
df.head()
```

```python
m_flx = et_b.creat_struct_saida(np.array(df["macho_p_imobilidade_flx"]))

m_flx["prob"] =[0.00, 0.00]
m_vei = et_b.creat_struct_saida(np.array(df["macho_p_imobilidade_veiculo"]))
m_vei["prob"] =[1, 1]


f_flx = et_b.creat_struct_saida(np.array(df["femea_p_imobilidade_flx"]))
f_flx["prob"] =[2, 2]

f_vei = et_b.creat_struct_saida(np.array(df["femea_p_imobilidade_veiculo"]))
f_vei["prob"] =[3, 3]

f_dif_v_t = et_b.creat_struct_saida(np.array(df["femea_dif_p"]))
m_dif_v_t = et_b.creat_struct_saida(np.array(df["macho_dif_p"]))

```

```python
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
```

```python
# figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
fig, axs = plt.subplots(2,1, figsize = (14,12))

sns.distplot(m_flx["data"],hist_kws={'weights': m_flx["x_weights"]}, kde=False, label="Macho tratado com Fluoxetina", ax= axs[0]);




sns.distplot(m_vei["data"],
             hist_kws={'weights': m_vei["x_weights"]}, 
             kde=False, label="Macho controle", ax= axs[0]);




sns.distplot(f_flx["data"],
             hist_kws={'weights': m_flx["x_weights"]}, 
             kde=False, label="Femea tratada com Fluoxetina", ax= axs[0]);



sns.distplot(f_vei["data"],
             hist_kws={'weights': m_vei["x_weights"]}, 
             kde=False, label="Femea controle", ax= axs[0]);


# sns.lineplot(x="hdi", y="prob", data=m_vei, label="HDI Macho Veiculo")
# sns.lineplot(x="hdi", y="prob", data=f_flx, label="HDI Femea Fluoxetina")
# sns.lineplot(x="hdi", y="prob", data=m_flx, label="HDI Macho Fluoxetina")
# sns.lineplot(x="hdi", y="prob", data=f_vei, label="HDI Macho Veiculo")


left, right = axs[0].get_xlim()
axs[1].set_xlim((left, right))
sns.lineplot(x="hdi", y="prob", data=m_flx, label="HDI Macho Fluoxetina", ax= axs[1])
sns.lineplot(x="hdi", y="prob", data=m_vei, label="HDI Macho controle", ax= axs[1])

sns.lineplot(x="hdi", y="prob", data=f_flx, label="HDI Fêmea Fluoxetina", ax= axs[1])
sns.lineplot(x="hdi", y="prob", data=f_vei, label="HDI Fêmea controle", ax= axs[1])


axs[0].set_title("Distribuição de probabilidade a posteriori da probabilidade de imobilidade")
axs[1].set_title("Intervalo de credibilidade de 94%")

axs[0].set_ylabel("Probabilidade")
# axs[1].set_ylabel("Probabilidade")
axs[1].set_yticklabels([])



axs[1].set_xlabel("Probabilidade da expressão de imobilidade a cada segundo de análise")

axs[0].legend()

plt.savefig('dist_prob_intevalo.png')
# plt.show()

# fmri = {}
# fmri["hdi"] = az.hdi(data, hdi_prob=.94)
# fmri["prob"] =[0.01, 0.01]
```

```python
figure(num=None, figsize=(14, 7), dpi=80, facecolor='w', edgecolor='k')
# fig, axs = plt.subplots(1,1, figsize = (12,12))

sns.distplot(m_flx["data"],hist_kws={'weights': m_flx["x_weights"]}, kde=False, label="Macho tratado com Fluoxetina")#, ax= axs[0]);




sns.distplot(m_vei["data"],
             hist_kws={'weights': m_vei["x_weights"]}, 
             kde=False, label="Macho controle")#, ax= axs[0])#;




sns.distplot(f_flx["data"],
             hist_kws={'weights': m_flx["x_weights"]}, 
             kde=False, label="Fêmea tratada com Fluoxetina")#, ax= axs[0]);



sns.distplot(f_vei["data"],
             hist_kws={'weights': m_vei["x_weights"]}, 
             kde=False, label="Fêmea controle")#, ax= axs[0]);


# sns.lineplot(x="hdi", y="prob", data=m_vei, label="HDI Macho Veiculo")
# sns.lineplot(x="hdi", y="prob", data=f_flx, label="HDI Femea Fluoxetina")
# sns.lineplot(x="hdi", y="prob", data=m_flx, label="HDI Macho Fluoxetina")
# sns.lineplot(x="hdi", y="prob", data=f_vei, label="HDI Macho Veiculo")

dif_m = et_b.data_saida(m_vei["data"],m_flx["data"])
dif_m["prob"] = [0.01, 0.01]
dif_f = et_b.data_saida(f_vei["data"],f_flx["data"])
dif_f["prob"] = [0.05, 0.05]

dif_mtra_fcon = et_b.data_saida(m_flx["data"],f_vei["data"])
dif_mtra_fcon["prob"] = [0.065, 0.065]


dif_mcon_ftra = et_b.data_saida(m_vei["data"],f_flx["data"])
dif_mcon_ftra["prob"] = [0.065, 0.065]
# left, right = axs[0].get_xlim()
# axs[1].set_xlim((left, right))
sns.lineplot(x="hdi", y="prob", data=dif_m)#, label="HDI Macho Fluoxetina")#, ax= axs[1])
sns.lineplot(x="hdi", y="prob", data=dif_f)#, label="HDI Macho Fluoxetina")#, ax= axs[1])
sns.lineplot(x="hdi", y="prob", data=dif_mtra_fcon)#, label="HDI Macho Fluoxetina")#, ax= axs[1])
sns.lineplot(x="hdi", y="prob", data=dif_mcon_ftra)#, label="HDI Macho Fluoxetina")#, ax= axs[1])



plt.text(0.27, 0.053, f'|d|={dif_f["cohen_d"]:.2f}', horizontalalignment='center', verticalalignment='center')
plt.text(0.25, 0.012, f'|d|={dif_m["cohen_d"]:.2f}', horizontalalignment='center', verticalalignment='center')

plt.text(0.24, 0.067, f'|d|={dif_mtra_fcon["cohen_d"]:.2f}', horizontalalignment='center', verticalalignment='center')

plt.text(0.30, 0.067, f'|d|={dif_mcon_ftra["cohen_d"]:.2f}', horizontalalignment='center', verticalalignment='center')

# sns.lineplot(x="hdi", y="prob", data=m_vei, label="HDI Macho controle", ax= axs[1])

plt.title("Distância entre Distribuições a posteriori de Probabilidade do comportamento de Imobilidade\n de Machos e Femêas tratados e seus controles")
plt.ylabel("Probabilidade")
plt.xlabel("Probabilidade da expressão de Imobilidade a cada segundo de análise")
# sns.lineplot(x="hdi", y="prob", data=f_flx, label="HDI Femea Fluoxetina", ax= axs[1])
# sns.lineplot(x="hdi", y="prob", data=f_vei, label="HDI Femea controle", ax= axs[1])
# plt.title("Distribuição a posteriori da probabilidade de expressão do comportamento de imobilidade de machos no TNF")
# plt.ylabel("Distribuição da probabilidade de expressão da imobilidade")
# plt.xlabel("Probabilidade da expressão de imobilidade a cada segundo de análise")
plt.legend()
# plt.show()
plt.savefig('dist_poste_cohen_d.png')

# fmri = {}
# fmri["hdi"] = az.hdi(data, hdi_prob=.94)
# fmri["prob"] =[0.01, 0.01]
```

```python
figure(num=None, figsize=(14, 7), dpi=80, facecolor='w', edgecolor='k')
sns.distplot(m_dif_v_t["data"],
             hist_kws={'weights': m_flx["x_weights"]}, 
             kde=False, label="Macho");


sns.distplot(f_dif_v_t["data"],
             hist_kws={'weights': m_vei["x_weights"]}, 
             kde=False, label="Fêmea");




dif_m_f = et_b.data_saida(m_dif_v_t["data"], f_dif_v_t["data"])
dif_m_f["prob"] = [0.01, 0.01]


sns.lineplot(x="hdi", y="prob", data=dif_m_f)#, label="HDI Macho Fluoxetina")#, ax= axs[1])
plt.text(0, 0.014, f'|d|={dif_m_f["cohen_d"]:.2f}', horizontalalignment='center', verticalalignment='center')



p_maior_macho = et_b.calc_p_menor_zero(m_dif_v_t["data"])
p_menor_femea = et_b.calc_p_maior(f_dif_v_t["data"])

sns.lineplot(x="hdi", y="prob", data=p_maior_macho)#, label="HDI Macho Fluoxetina")#, ax= axs[1])
plt.text(-0.1, 0.003 , f'p={p_maior_macho["p"]:.2f}', horizontalalignment='center', verticalalignment='center')


sns.lineplot(x="hdi", y="prob", data=p_menor_femea)#, label="HDI Macho Fluoxetina")#, ax= axs[1])
plt.text(0.1, 0.063 , f'p={p_menor_femea["p"]:.2f}', horizontalalignment='center', verticalalignment='center')



plt.title("A diferença entre a Distribuição a Posteriori da Probabilidade de Imobilidade \n do controle e do tratamento de machos e femêas \n")
plt.ylabel("Probabilidade")
plt.xlabel("Diferença de probabilidade da expressão de imobilidade a cada segundo de análise entre controle e tratamento")
plt.legend()
# plt.show()
plt.savefig('dif_dis_cohen.png')
```
